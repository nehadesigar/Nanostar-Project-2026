```python
import time
import os
```

```python
print("Stopwatch started. Press Ctrl+C to stop.")
start_time = time.perf_counter()
log_dir = "sim_logs"
os.makedirs(log_dir, exist_ok=True)
progress_file = os.path.join(log_dir, "time trial.txt")


try:
    while True:
        # Calculate elapsed time
        time_elapsed = time.perf_counter() - start_time
        
        
        # Format as minutes and seconds
        mins = time_elapsed // 60
        secs = time_elapsed % 60
        
        # Print with \r to overwrite the line in the console
        with open(progress_file, "w") as f:
            f.write(f"time={mins:.0f}:{secs:.0f}")
        
        # Wait for exactly 1 second
        time.sleep(1)

except KeyboardInterrupt:
    print("\nStopwatch stopped.")
```
